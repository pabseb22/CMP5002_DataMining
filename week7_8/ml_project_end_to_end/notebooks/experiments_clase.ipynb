```python
lr = 0.1
n_epochs = 1000

m = len(X_b)
```

```python
# Stochastic Gradient Descent
```

```python
n_epochs_sgd = 50

#Hiperparametros del decay del LR
t0, t1 = 5, 50

theta_sgd = np.random.randn(2,1)
```

```python
def learning_schedule(t):
    return t0 / (t + t1)
```

```python
for epoch in range(1, n_epochs_sgd):

    for iteracion in range(m):

        random_index = np.random.randint(m)
        xi = X_b[random_index: random_index + 1]
        yi = y[random_index: random_index + 1]

        gradient_vector_sdg = 2 * xi.T @ (xi @ theta_sgd - yi)

        lr = learning_schedule(iteracion + m * epoch)

        theta_sgd = theta_sgd - lr * gradient_vector_sdg
```

```python
theta_sgd
```

```python
# Dummy Mini Batch SGD
# Que en cada iteracion se calcula el vector de gradiente con un subconjunto de los datos 
```

```python
mini_batch_max = 10
theta_mini_sgd = np.random.randn(2,1)

for epoch in range(1, n_epochs_sgd):

    for iteracion in range(m):

        random_index = np.random.randint(m)

        # Validacion para el indice no exceda el numero de muestras
        if random_index + mini_batch_max > m:
            random_index = m - mini_batch_max
            xi = X_b[random_index: random_index + mini_batch_max]
            yi = y[random_index: random_index + mini_batch_max]
        else: # caso normal
            xi = X_b[random_index: random_index + mini_batch_max]
            yi = y[random_index: random_index + mini_batch_max]

        # tamaño del mini batch
        batch_size = xi.shape[0]

        gradient_vector_mini_sdg = (2 / batch_size) * xi.T @ (xi @ theta_mini_sgd - yi)

        lr = learning_schedule(iteracion + m * epoch)

        theta_mini_sgd = theta_mini_sgd - lr * gradient_vector_mini_sdg
theta_mini_sgd
```
